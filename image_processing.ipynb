```python
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('pup.jpg')
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.imread('pup2.jpg')
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
print(img1.shape)
print(img2.shape)
```

## Blending imges of same shape

```python
img1 = cv2.resize(img1,(1000,1000))
img2 = cv2.resize(img2,(1000,1000))
```

```python
plt.imshow(img1)
```

```python
blended = cv2.addWeighted(src1=img1,alpha=0.8,src2=img2,beta=0.7,gamma=1)
```

```python
plt.imshow(blended)
```

## Overlay small images on top of larger images (no blending)

```python
img1 = cv2.imread('pup.jpg')
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.imread('pup2.jpg')
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
img1 = cv2.resize(img1,(600,600))
```

```python
large_img = img1
small_img = img2
print(small_img.shape)
```

```python
plt.imshow(large_img)
```

```python
x_offset = 0
y_offset = 0
```

```python
x_end = x_offset + small_img.shape[1]
y_end = y_offset + small_img.shape[0]
```

```python
large_img[y_offset:y_end,x_offset:x_end] = small_img
```

```python
plt.imshow(large_img)
```
